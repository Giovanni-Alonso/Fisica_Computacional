## Ejercicio 1

El método de Euler sirve para obtener una solución aproximada de la ecuación diferencial:

$$\frac{dx}{dt}=f(x,t)$$

Implementando la fórmula de recurrencia siguiente:

$$x_{k+1}=x_{k}+hf(x_{k},t_{k}),$$

en donde $x_{k}$ es la aproximación de $x(t_{k})$ y $t_{k}=t_{0}+hk$. Como todo ecuación diferencial de primer orden necesitamos una condición inicial, la cual está dada por $x_{0}=x(t_{0})$.

```julia
function metodo_euler(f,x0,t0,tf,h)
    
    t = t0
    x = x0
    listt = []
    listx = []
    push!(listt,t)
    push!(listx,x)
    n = round((tf-t0)/h)
    for i in 1:n-1
        x += h*f(x,t)
        t += h
        push!(listt,t)
        push!(listx,x)
    end
    
    return listt, listx
    
end;
```

Con `metodo_euler` vamos a obtener solución a la ecuación diferencial:

$$\frac{dx}{dt}=-2.5x, \: x_{0}=x(0)=1\text{ y }2, \text{ y } t\in(0,8),$$

para los pasos $h=1.0,0.5,0.2,0.01$. Para observar la convergencia del método se comparará con la solución exacta que es:

$$x(t)=x_{0}e^{-2.5t}$$

```julia
f(x,t)=-2.5x;
```

```julia
S11=metodo_euler(f,1,0,8,1);
```

```julia
scatter(S11[1],S11[2],label=L"Método de Euler ($h=1$)",s=10)
plot(S11[1],S11[2])
t = linspace(0,8,100)
plot(t,exp.(-2.5*t),label=L"Solución exacta ($x(0)=1$)")
legend()
show();
```

```julia
S12=metodo_euler(f,1,0,8,0.5);
S13=metodo_euler(f,1,0,8,0.2);
S14=metodo_euler(f,1,0,8,0.01);
```

```julia
scatter(S12[1],S12[2],label=L"Método de Euler ($h=0.5$)",s=10)
plot(S12[1],S12[2])
scatter(S13[1],S13[2],label=L"Método de Euler ($h=0.2$)",s=10)
plot(S13[1],S13[2])
scatter(S14[1],S14[2],label=L"Método de Euler ($h=0.01$)",s=5)
plot(S14[1],S14[2])
t = linspace(0,8,100)
plot(t,exp.(-2.5*t),label=L"Solución exacta ($x(0)=1$)")
legend()
show();
```
